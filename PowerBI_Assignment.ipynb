
## Question 1  
**Define Power BI and the key components of the Power BI ecosystem**

**Answer:**  
Power BI is a business analytics tool by Microsoft that helps users visualize data,
create interactive reports, and share insights across an organization.

**Key Components:**
- **Power BI Desktop:** Windows-based tool used for data modeling, DAX, and report creation.
- **Power BI Service:** Cloud platform for publishing, sharing, and collaborating on reports and dashboards.
- **Power BI Mobile:** Mobile apps for Android/iOS to access dashboards on the go.
- **Power BI Gateway:** Acts as a bridge between on‑premises data sources and Power BI Service.



## Question 2  
**Pie Chart vs Donut Chart | Bar Chart vs Column Chart**

**Answer:**  
- **Pie vs Donut:** Both show part-to-whole. Donut allows center labels/KPIs.  
  *Example:* Market share by region → Donut chart.

- **Bar vs Column:** Bar is horizontal (good for long labels), Column is vertical (time-based).  
  *Example:* Sales by Sub‑Category → Bar chart.



## Question 3  
**Star Schema vs Snowflake Schema | Primary Key vs Foreign Key**

**Answer:**  
- **Star Schema:** Central fact table with dimension tables – faster performance.
- **Snowflake Schema:** Normalized dimensions – complex but storage efficient.

- **Primary Key:** Unique identifier in a table.
- **Foreign Key:** Links fact to dimension tables.

**Cardinality Importance:**  
Defines relationship behavior (One-to-Many) and ensures accurate aggregations.



## Question 4  
**Calculated Column vs Measure | Row Context vs Filter Context**

**Answer:**  
- **Calculated Column:** Computed at row level, stored in table.
- **Measure:** Calculated at runtime based on filters.

- **Row Context:** Calculation per row (e.g., Profit = Sales - Cost).
- **Filter Context:** Filters applied by visuals/slicers.



## Question 5  
**Difference between Report and Dashboard**

**Answer:**  
- **Report:** Multi-page, interactive, built in Power BI Desktop.
- **Dashboard:** Single-page summary created in Power BI Service using pinned visuals.

**Step 2: Calculate Total Sales (Manual)**

**Step 3: Percentage Formula**

- Percentage of Total Sales =
(Category Sales ÷ Total Sales) × 100

2️⃣ Power BI – DAX Measure

Create this measure in Power BI Desktop → Modeling → New Measure

% of Total Sales = DIVIDE(
    SUM('Product_Sales'[Sales_Amount]),
    CALCULATE(
        SUM('Product_Sales'[Sales_Amount]),
        ALL('Product_Sales'[Product_Category])
    )
)

To calculate the percentage of total sales by product category, a DAX measure was created using the DIVIDE() and ALL() functions. The measure sums the sales amount for each category and divides it by the total sales across all categories. The ALL() function removes category-level filters to ensure that the denominator always represents total sales.

This measure enables dynamic calculation of sales contribution regardless of filters applied in the report. The results were displayed using a table visual and a donut chart in Power BI. The table provides a precise numerical breakdown, while the donut chart visually highlights high-contributing categories such as Tables & Chairs and Home Appliances. This analysis helps management identify key revenue-driving product categories and supports data-driven business decisions.
